### 1er grafico

En la siguiente visualización se presenta la comparación de algunos atributos importantes (edad, genero, educación, etc.) de los usuarios al año 2010 de dos de las redes sociales más populares, Facebook Vs Twitter.


+ **Abstracción del ¿qué?, ¿por qué? y ¿cómo?**

	- What? = Tabla con Ítems representados por los usuarios y Atributos de tipo categóricos como género, educación y ubicación; y de tipo cuantitativo como edad e ingreso que para efectos de la visualización fueron llevados a rangos.
	- ¿Why? = Present y Compare features
	- ¿How? = Separate-> Facet(partition) -> Navigate -> Arrange


+ **Marcas y canales usados**

    - Marca de áreas
    - Canal de color para cada una de las categorías de los atributos
    
    
+ **¿Cumple o no cumple alguna de las reglas generales descritas?**

No cumple la regla de “Primero la función, luego la forma”. Se debe priorizar la funcionalidad antes que la estética en una visualización, y el grafico anterior, representado como una especie de pie chart irregular que muestra muchos valores y categorías, aunque visualmente sea muy llamativo no es muy efectivo, dado que no expresa concretamente la función del gráfico, no permite facilidad en la percepción y entendimiento intuitivo de lo que se intenta reflejar. 


+ **Mejoras posibles, en cuánto a marcas o canales, que le realizaría a la visualización**

Se me ocurre la elaboración de gráficas de barras agrupadas(marca de áreas) para los atributos presentados como un solo porcentaje y tablas para los atributos con varias categorías, esto por supuesto discriminado por red social(canal de color para atributo "tipo de red social"), que permita comprender al espectador con mayor detalle la funcionalidad de la visualización.


**Fuente:** <https://www.digitalsurgeons.com/thoughts/creative/social-demographics-2010-a-fresh-look-at-facebook-and-twitter/>

### 2do grafico


La siguiente grafica nos muestra las películas más exitosas del 2008 al 2016 y algunos de sus atributos (genero, presupuesto, criticas, entre otros)

+ **Abstracción del ¿qué?, ¿por qué? y ¿cómo?**

	- What? = Tabla con Ítems representados por las películas y Atributos de tipo temporal siendo representado por los años de estreno; de tipo categórico como género o tipo de película; y de tipo cuantitativo como el presupuesto recuperado o el promedio de críticas
	- ¿Why? = Present, Enjoy, Look up correlation y trend
	- ¿How? = Encode -> Navigate -> Select -> Superimpose -> Filter -> Aggregate


+ **Marcas y canales usados**

    - Marca de puntos
    - Canal de posición espacial horizontal para el atributo cuantitativo (Promedio de críticas)
    - Canal de posición espacial vertical para el atributo cuantitativo (Presupuesto recuperado)
    - Canal de color para el atributo categórico (Género)
    - Canal de tamaño para el atributo cuantitativo (Presupuesto)

Lo anterior varía de acuerdo al filtro realizado en las opciones “Colour”, “Size” y “X-Axis” y “Y-Axis”
    
+ **¿Cumple o no cumple alguna de las reglas generales descritas?**

Cumple la regla del mantra de la visualización, “Primero overview, zoom y filtro, detalles a petición”, ya que permite al espectador la experiencia de navegar y personalizar mediante filtros por distintos criterios, ya sea por genero de película, por presupuesto, por ganancias, entre otros atributos. Además, tiene la interactiva opción de que al dar clic sobre alguna película dirige al espectador a la página de Amazon Prime Video dónde se podrá ver dicha película. Adicionalmente, cumple con la regla de "Se requiere capacidad de respuesta", dado que al cambiar cualquiera de los parámetros, en especial el de color, la transición al nuevo diseño genera una respuesta perceptiva inmediata.


+ **Mejoras posibles, en cuánto a marcas o canales, que le realizaría a la visualización**

No considero que la visualización debiera tener mejoras, debido a que tal cual como se encuentra proporciona gran información y de forma muy detalladas y personalizada para el espectador, que hace que sea posible la comprensión intuitiva e inmediata de lo expresado allí.

**Fuente:** <https://www.informationisbeautiful.net/visualizations/the-hollywood-insider/>
